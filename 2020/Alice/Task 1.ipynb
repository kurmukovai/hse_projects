```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
vgg_acc = np.load('vgg_acc.npy', allow_pickle=True)
mob_acc = np.load('mob_acc.npy', allow_pickle=True)
vgg16_acc = np.load('vgg16_acc.npy', allow_pickle=True)
```

```python
plt.rcParams["figure.figsize"] = (6, 5)
plt.title('Accuracy')
plt.xlabel('epoch', fontsize=13)
plt.scatter(np.arange(1, 51), vgg_acc, label='VGG custom');
plt.scatter(np.arange(1, 51), vgg16_acc, label='VGG-16');
plt.scatter(np.arange(1, 51), mob_acc, label='MobileNetv2');
plt.legend();
```
